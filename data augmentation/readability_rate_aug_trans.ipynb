```python
import numpy as np 
import pandas as pd 
import os
from textblob import TextBlob
import time
```

```python
df = pd.read_csv("../input/commonlitreadabilityprize/train.csv")
df.head()
```

```python
df['text'] = [''] * len(df)

for index, row in df.iterrows():
    italian_translation = str(TextBlob(row['excerpt']).translate(to='it'))
    df.loc[index,'text'] = str(TextBlob(italian_translation).translate(to='en'))
    time.sleep(10)
    
df.head()
```
